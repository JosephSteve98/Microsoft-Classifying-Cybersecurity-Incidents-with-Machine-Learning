```python
null_percent = df_test.isna().mean()*100
columns_to_drop = null_percent[null_percent > 50].index
columns_to_drop
```

```python
df_test_clean = df_test.drop(columns = columns_to_drop)
df_test_clean.shape
```

```python
df_test_clean.isna().sum()
```

```python
df_test_clean.nunique()
```

```python
df_test_clean['Timestamp'] = pd.to_datetime(df_test_clean['Timestamp'])
```

```python
df_test_clean['Timestamp'].head()
```

```python
# Extract day, month, and hour from the Timestamp
df_test_clean['Day'] = df_test_clean['Timestamp'].dt.day
df_test_clean['Month'] = df_test_clean['Timestamp'].dt.month
df_test_clean['Hour'] = df_test_clean['Timestamp'].dt.hour
df_test_clean['Year'] = df_test_clean['Timestamp'].dt.year


# Drop the original Timestamp column
df_test_clean.drop('Timestamp', axis=1, inplace=True)

print(df_test_clean.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='IncidentGrade', data=df_test_clean)
plt.title('Distribution of Incident Grades')
plt.show()
```

```python
# Aggregate data
hour_incident_grade = df_test_clean.groupby(['Hour', 'IncidentGrade']).size().unstack()

# Plot
hour_incident_grade.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.xlabel('Hour of Day')
plt.ylabel('Count')
plt.title('IncidentGrade Counts Across Hours of the Day')
plt.legend(title='IncidentGrade')
plt.grid(True)
plt.show()
```

```python
# Aggregate data
day_incident_grade = df_test_clean.groupby(['Day', 'IncidentGrade']).size().unstack()

# Plot
day_incident_grade.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.xlabel('Day of the Month')
plt.ylabel('Count')
plt.title('IncidentGrade Counts Across Days of the Month')
plt.legend(title='IncidentGrade')
plt.grid(True)
plt.show()
```

```python
# Aggregate data
month_incident_grade = df_test_clean.groupby(['Month', 'IncidentGrade']).size().unstack()

# Plot
month_incident_grade.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.xlabel('Month of Year')
plt.ylabel('Count')
plt.title('IncidentGrade Counts Across Months')
plt.legend(title='IncidentGrade')
plt.grid(True)
plt.show()
```

```python
# Select numerical columns
numerical_df = df_test_clean.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numerical_df.corr().abs()

# Plot the heatmap
plt.figure(figsize=(20,17))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Setting a correlation threshold
threshold = 0.8

# Find pairs of columns with high correlations
high_corr_var = np.where(corr_matrix > threshold)
high_corr_pairs = [(numerical_df.columns[x], numerical_df.columns[y])
                    for x, y in zip(*high_corr_var)
                    if x != y and x < y]

# Create a list of columns to remove
columns_to_remove = set()
for col1, col2 in high_corr_pairs:
    columns_to_remove.add(col2)  # or col1, based on your strategy

# Convert to list
columns_to_remove = list(columns_to_remove)

#print(columns_to_remove)
# Drop the redundant columns
df_cleaned_corr = df_test_clean.drop(columns=columns_to_remove)

# Verify the cleaned DataFrame
print("Columns removed:", columns_to_remove)
print(df_cleaned_corr.head())
```

```python
from sklearn.feature_selection import f_classif
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from scipy.stats import pearsonr

categorical_cols = df_cleaned_corr.select_dtypes(include=['object']).columns

# Initialize a dictionary to hold the label encoders
label_encoders = {}

# Apply LabelEncoder to each categorical column
for column in categorical_cols:
    le = LabelEncoder()
    df_cleaned_corr[column] = le.fit_transform(df_cleaned_corr[column])
    label_encoders[column] = le

# The DataFrame now has label-encoded categorical columns and any numerical columns
print(df_cleaned_corr)
```

```python
import joblib

# Path to save the label encoders
label_encoders_path = "label_encoders.pkl"

# Save the label encoders dictionary to a file
joblib.dump(label_encoders, label_encoders_path)

print(f"Label encoders saved to {label_encoders_path}")
```
